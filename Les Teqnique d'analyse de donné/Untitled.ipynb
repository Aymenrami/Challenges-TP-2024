```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from scipy.stats import stats
```

```python
data = {
    'ville': ['casa', 'agadir', 'tanger', 'casa', 'fes', 'oujda', 'oujda', 'casa', 'rabat', 'tanger', 'tanger', 'nador', 'casa', 'fes', 'casa', 'rabat'],
    'age': [25, 27, 29, 26, 24, 28, 27, 30, 31, 32, 29, 28, 28, 26, None, None]
}

df = pd.DataFrame(data)
df
```

```python
plt.style.use('dark_background')
plt.figure(figsize = (5,4))
sns.histplot(data=df, x='age', kde=True, color='#00bbff')
plt.title('Histogram of Age')
plt.xlabel('Age', fontsize=12, color='white')
plt.ylabel('') 
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.show()
```

```python
messing = df.isna().sum()
messing
```

```python
missing = df.isna().sum()
total = df.shape[0]  
pourcentage = ( missing / total ).round(2)*100
result_table = pd.DataFrame({
    'Nombre de valeurs manquantes': missing,
    'Fréquence des valeurs manquantes (%)':pourcentage
})
result_table
```

# le Pourcentage est >15%

```python
mean_age = df['age'].mean()
df['age'] = df['age'].fillna(mean_age)
df.isna().sum()
```

```python
plt.style.use('dark_background')
plt.figure(figsize = (5,4))
sns.histplot(data=df, x='age', kde=True, color='#00bbff')
plt.title('Histogram of Age')
plt.xlabel('Age', fontsize=12, color='white')
plt.ylabel('') 
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.show()
```

```python
mean_age = df['age'].mean()
df['age'].fillna(mean_age, inplace=True)
print(df)

```

```python
plt.figure(figsize=(5, 3))
stats.probplot(df['age'], dist="norm", plot=plt)
plt.title('QQ Plot using Scipy')
plt.style.use('dark_background')
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.show()
```
