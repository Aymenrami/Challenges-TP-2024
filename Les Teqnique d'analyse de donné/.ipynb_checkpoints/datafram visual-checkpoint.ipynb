```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data = {
    'Mois': ['Janvier', 'Février', 'Mars', 'Avril', 'Mai', 'Juin', 'Juillet', 'Août', 'Septembre', 'Octobre', 'Novembre', 'Décembre'],
    'Produit A': [200.0, 220.0, None, 250.0, 270.0, 300.0, 310.0, 330.0, None, 360.0, 380.0, 400.0],
    'Produit B': [150.0, 160.0, 170.0, None, 190.0, 200.0, 210.0, 220.0, 230.0, 240.0, None, 260.0],
    'Produit C': [100.0, None, 120.0, 130.0, 140.0, 150.0, None, 170.0, 180.0, 190.0, 200.0, None]
}

df = pd.DataFrame(data)

```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
missing = df.isna().sum()
total = df.shape[0]  
pourcentage = ( missing / total ).round(2)*100
result_table = pd.DataFrame({
    'Nombre de valeurs manquantes': missing,
    'Fréquence des valeurs manquantes (%)':pourcentage
})
result_table
```

```python
sns.barplot(data=missing)
```

```python
sns.heatmap(data = result_table)
```

```python
sns.heatmap(df.isnull(),cbar=False,cmap='rocket')
```
