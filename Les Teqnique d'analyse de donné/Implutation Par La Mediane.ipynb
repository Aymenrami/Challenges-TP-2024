```python
import pandas as pd
from scipy.stats import stats
data = {
    'ville': ['casa', 'agadir', 'tanger', 'casa', 'fes', 'oujda', 'oujda', 'casa', 'rabat', 'tanger', 'tanger', 'nador', 'casa', 'fes', 'casa', 'rabat'],
    'age': [25, 27, 29, 26, 24, 28, 27, 30, 31, 32, 29, 28, 28, 26, None, None]
}
df = pd.DataFrame(data)
df
```

```python
def ecart_type(data):
    moyenne = sum(data) / len(data)
    variance = sum((x - moyenne) ** 2 for x in data) / len(data)
    ecart_type = variance ** 0.5
    return ecart_type
```

```python
def Skewness(Data):
    moyenne = sum(Data) / len(Data)
    ecart_type_value = ecart_type(Data)
    n = len(Data)
    Skew1 = n / ((n - 1) * (n - 2))
    Skew2 = sum(((x - moyenne) / ecart_type_value) ** 3 for x in Data)
    skewf = Skew1 * Skew2
    return skewf
```
